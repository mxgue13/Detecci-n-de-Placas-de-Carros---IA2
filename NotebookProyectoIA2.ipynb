```python
import torch
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os
import random
%matplotlib inline
# Inicializar lector EasyOCR
reader = easyocr.Reader(['en'])

# Cargar modelo YOLOv5 entrenado
model_path = '/content/drive/MyDrive/IA2/runs/detect/trainv5/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

# Carpeta de imÃ¡genes de prueba
test_folder = '/content/drive/MyDrive/IA2/data/images/test/'
all_images = [f for f in os.listdir(test_folder) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
random_image = random.choice(all_images)
img_path = os.path.join(test_folder, random_image)

# Inferencia con YOLOv5
results = model(img_path)

# Obtener cajas (formato xyxy)
boxes = results.xyxy[0].cpu().numpy()
img = cv2.imread(img_path)

# Iterar sobre cada placa detectada
for i, box in enumerate(boxes):
    x1, y1, x2, y2, conf, cls = box.astype(int)
    placa = img[y1:y2, x1:x2]

    # Preprocesamiento
    gray = cv2.cvtColor(placa, cv2.COLOR_BGR2GRAY)
    gray = cv2.bilateralFilter(gray, 11, 17, 17)

    # OCR con EasyOCR
    results_ocr = reader.readtext(gray)

    # Mostrar todos los textos detectados
    if results_ocr:
        print(f'Placa {i+1}:')
        for j, (bbox, text, conf) in enumerate(results_ocr):
            print(f'  Texto {j+1}: "{text}" (confianza: {conf:.2f})')
    else:
        print(f'Placa {i+1}: No se detectÃ³ texto')

    # Mostrar imagen recortada de la placa
    plt.imshow(cv2.cvtColor(placa, cv2.COLOR_BGR2RGB))
    plt.title(f'Placa {i+1}: {text}')
    plt.axis('off')
    plt.show()

```

```python
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import torch
import easyocr

# Inicializar lector de EasyOCR
reader = easyocr.Reader(['en'])

# Cargar modelo YOLOv5 entrenado
model_path = '/content/drive/MyDrive/IA2/runs/detect/trainv5/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

# Ruta a la carpeta de test
VAL_DIR = '/content/drive/MyDrive/IA2/data/images/test/'

# Ruta donde guardar el CSV
CSV_OUTPUT = '/content/drive/MyDrive/IA2/resultados_placas_test_final_Yolov5.csv'

# Lista de imÃ¡genes
image_files = [f for f in os.listdir(VAL_DIR) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]

# Lista para almacenar resultados
results_list = []

# Iterar por cada imagen
for img_name in image_files:
    img_path = os.path.join(VAL_DIR, img_name)
    img = cv2.imread(img_path)

    # Inferencia con YOLOv5
    results = model(img_path)
    boxes = results.xyxy[0].cpu().numpy()

    # Recorrer cada placa detectada
    for i, box in enumerate(boxes):
        x1, y1, x2, y2, conf, cls = box.astype(int)
        placa_crop = img[y1:y2, x1:x2]

        # Preprocesamiento (puedes ajustar si quieres mejorar OCR)
        gray = cv2.cvtColor(placa_crop, cv2.COLOR_BGR2GRAY)
        gray = cv2.bilateralFilter(gray, 11, 17, 17)

        # OCR con EasyOCR
        ocr_result = reader.readtext(gray)

        if ocr_result:
            texto = ocr_result[0][1]  # Tomar el texto mÃ¡s confiable (primero)
        else:
            texto = ''

        # Guardar resultado
        results_list.append({
            'imagen': img_name,
            'n_placa': f'placa_{i+1}',
            'texto': texto
        })

# Guardar resultados en CSV
df = pd.DataFrame(results_list)
df.to_csv(CSV_OUTPUT, index=False)

print(f"âœ… CSV guardado en: {CSV_OUTPUT}")
df.head()
```

#PRUEBA CON EL DATASET DE PLACAS COLOMBIANAS EASYOCR

```python
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import torch
import easyocr
import re

# Inicializar modelo YOLOv5
model_path = '/content/drive/MyDrive/IA2/runs/detect/trainv5/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

# Inicializar lector OCR
reader = easyocr.Reader(['en'])

# Rutas
VAL_DIR = '/content/drive/MyDrive/IA2/DatasetTesseract'
CSV_OUTPUT = '/content/drive/MyDrive/IA2/resultados_placas_test_colombiaEasy.csv'

# Preparar imÃ¡genes
image_files = [f for f in os.listdir(VAL_DIR) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
results_list = []
total_chars = 0
correct_chars = 0

for img_name in image_files:
    img_path = os.path.join(VAL_DIR, img_name)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Ground truth desde nombre de archivo
    expected_plates = os.path.splitext(img_name)[0].upper().split('_')

    results = model(img_path)
    boxes = results.xyxy[0].cpu().numpy()

    detected_plates = []

    for i, box in enumerate(boxes):
        x1, y1, x2, y2, conf, cls = box.astype(int)

        # Recorte de la placa
        placa_crop = img[y1:y2, x1:x2]

        # Recortar bordes para evitar detecciones falsas como "I" del marco
        h, w = placa_crop.shape[:2]
        crop_recortado = placa_crop[int(0.05*h):int(0.95*h), int(0.05*w):int(0.95*w)]

        # OCR
        ocr_result = reader.readtext(crop_recortado, detail=0, paragraph=True)
        raw_text = ''.join(ocr_result)

        # Limpieza: solo letras y nÃºmeros, en mayÃºsculas
        clean_text = re.sub(r'[^A-Z0-9]', '', raw_text.upper())[:6]

        # Dibujar caja y texto
        cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img_rgb, clean_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.9, (255, 0, 0), 2, cv2.LINE_AA)

        detected_plates.append(clean_text)

        # Guardar resultado para CSV
        results_list.append({
            'imagen': img_name,
            'n_placa': f'placa_{i+1}',
            'texto': clean_text
        })

    # Comparar carÃ¡cter a carÃ¡cter
    for pred, gt in zip(detected_plates, expected_plates):
        for p_char, g_char in zip(pred[:6], gt[:6]):
            total_chars += 1
            if p_char == g_char:
                correct_chars += 1

    # Mostrar imagen con resultados
    plt.figure(figsize=(10, 8))
    plt.imshow(img_rgb)
    plt.title(f"{img_name}\nDetectado: {', '.join(detected_plates)}\nEsperado: {', '.join(expected_plates)}")
    plt.axis('off')
    plt.show()

# Guardar CSV
df = pd.DataFrame(results_list)
df.to_csv(CSV_OUTPUT, index=False)

# Imprimir resultados
accuracy_char = correct_chars / total_chars if total_chars > 0 else 0
print(f"\nâœ… CSV guardado en: {CSV_OUTPUT}")
print(f"ğŸ”¤ Accuracy por carÃ¡cter: {accuracy_char*100:.2f}% ({correct_chars}/{total_chars} caracteres correctos)")

```

```python
import os
import cv2
import pandas as pd
import matplotlib.pyplot as plt
import torch
import pytesseract
import re

# Inicializar modelo YOLOv5
model_path = '/content/drive/MyDrive/IA2/runs/detect/trainv5/weights/best.pt'
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path, force_reload=True)

# Rutas
VAL_DIR = '/content/drive/MyDrive/IA2/DatasetTesseract'
CSV_OUTPUT = '/content/drive/MyDrive/IA2/resultados_placas_test_colombiaTesseract.csv'

# Preparar imÃ¡genes
image_files = [f for f in os.listdir(VAL_DIR) if f.lower().endswith(('.jpg', '.png', '.jpeg'))]
results_list = []
total_chars = 0
correct_chars = 0

for img_name in image_files:
    img_path = os.path.join(VAL_DIR, img_name)
    img = cv2.imread(img_path)
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # Ground truth desde nombre de archivo
    expected_plates = os.path.splitext(img_name)[0].upper().split('_')

    results = model(img_path)
    boxes = results.xyxy[0].cpu().numpy()

    detected_plates = []

    for i, box in enumerate(boxes):
        x1, y1, x2, y2, conf, cls = box.astype(int)

        # Recorte de la placa
        placa_crop = img[y1:y2, x1:x2]

        # Recortar bordes para evitar marcos
        h, w = placa_crop.shape[:2]
        crop_recortado = placa_crop[int(0.05*h):int(0.95*h), int(0.05*w):int(0.95*w)]

        # OCR con Tesseract
        ocr_result = pytesseract.image_to_string(crop_recortado, config='--psm 7')
        raw_text = ocr_result.strip()

        # Limpieza: solo letras y nÃºmeros, en mayÃºsculas
        clean_text = re.sub(r'[^A-Z0-9]', '', raw_text.upper())[:6]

        # Dibujar caja y texto
        cv2.rectangle(img_rgb, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img_rgb, clean_text, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX,
                    0.9, (255, 0, 0), 2, cv2.LINE_AA)

        detected_plates.append(clean_text)

        # Guardar resultado para CSV
        results_list.append({
            'imagen': img_name,
            'n_placa': f'placa_{i+1}',
            'texto': clean_text
        })

    # Comparar carÃ¡cter a carÃ¡cter
    for pred, gt in zip(detected_plates, expected_plates):
        for p_char, g_char in zip(pred[:6], gt[:6]):
            total_chars += 1
            if p_char == g_char:
                correct_chars += 1

    # Mostrar imagen con resultados
    plt.figure(figsize=(10, 8))
    plt.imshow(img_rgb)
    plt.title(f"{img_name}\nDetectado: {', '.join(detected_plates)}\nEsperado: {', '.join(expected_plates)}")
    plt.axis('off')
    plt.show()

# Guardar CSV
df = pd.DataFrame(results_list)
df.to_csv(CSV_OUTPUT, index=False)

# Imprimir resultados
accuracy_char = correct_chars / total_chars if total_chars > 0 else 0
print(f"\nâœ… CSV guardado en: {CSV_OUTPUT}")
print(f"ğŸ”¤ Accuracy por carÃ¡cter: {accuracy_char*100:.2f}% ({correct_chars}/{total_chars} caracteres correctos)")
```
